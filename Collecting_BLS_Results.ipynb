```python
results_path = '/Users/daxfeliz/Desktop/Mentoring_Teaching/SRMP_project/2024_SRMP/SRMP_Research/'
```

```python
def collect_BLS_results(targets,results_path):
    RPs=[]
    Ps=[]
    Durs=[]
    T0s=[]
    SDEs=[]

    for t in range(len(targets)):
        target=targets.iloc[t]
        ID = target['TIC ID']
        bls_result_filename = results_path+'TIC_'+str(ID)+'_BLS_result.csv'
        try:
            bls_result = pd.read_csv(bls_result_filename)
            RPs = np.append(RPs,bls_result['Planet Radius [Earth Radii]'])
            Ps = np.append(Ps,bls_result['Period [days]'])
            Durs= np.append(Durs,bls_result['Transit Duration [hours]'])
            T0s= np.append(T0s,bls_result['Transit Epoch [BTJD]'])
            SDEs= np.append(SDEs,bls_result['BLS SDE'])
        except FileNotFoundError as FE:
            print(FE,'\n No BLS result for TIC ',ID)
            RPs = np.append(RPs,np.nan)
            Ps = np.append(Ps,np.nan)
            Durs= np.append(Durs,np.nan)
            T0s= np.append(T0s,np.nan)
            SDEs= np.append(SDEs,np.nan)
    print('len check:',len(targets),len(RPs),len(Ps),len(Durs),len(T0s),len(SDEs))
    
#     output_df = pd.DataFrame({'Planet Radius (Earth Radii)':RPs,
#                              'Period (days)':Ps,
#                              'Transit Duration (hours)':Durs,
#                              'Transit Epoch [BTJD]':T0s,
#                              'Signal Detection Efficiency (SDE)':SDEs})
    
    output_df = pd.DataFrame({'Planet Radius (Earth Radii)':RPs,
                             'Period (days)':Ps,
                             'Transit Duration (hours)':Durs,
                             'Signal Detection Efficiency (SDE)':SDEs})    
    return output_df
```

```python
o=collect_BLS_results(targets,results_path)
```
